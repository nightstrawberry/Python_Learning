### Requests库的7个主要方法
requests.request(method, url, \**kwargs) 构造请求

requests.get(url, params=None, \**kwargs)  

requests.head(url, \**kwargs)

requests.post(url, data=None, json=None, \**kwargs)

requests.put(url, data=None, \**kwargs)

requests.patch(url, data=None, \**kwargs) 局部修改

requests.delete(url, \**kwargs) 

### r = requests.get(url,params=None,**Kwargs)
**kwargs:控制访问参数(12个):

重要：json,params,data,headers,

cookies,auth

files(向服务器提交文件),timeout(设定超时时间/s)

proxies:字典类型，设定访问代理服务器，增加登录认证

```python
# headers:模拟浏览器向服务器发起访问
hd = {'user-agent':'Chrome/10'}
r = requests.request('POST','http://python123.io/ws',headers=hd)
```

```python
# 发起GER请求时，设定timeout时间
r = requests.request('GET','http://python123.io/ws',timeout=5)
```

```python
# 增加代理服务器
pxs = {'http':'http://user:pass@10.10.10.1:11234',\
      'http':'http://10.10.10.1:4321'}
r = requests.request('GET','http://python123.io/w',proxies=pxs)
```

# 异常处理
    try:

        return r.text

        return ("产生异常")

## 实例1：京东商品爬取

```python
import requests
r = requests.get("http://item.jd.com/2967929.html")
r.status_code
```

```python
r.encoding
```

```python
r.apparent_encoding
```

```python
r.text
```

```python
# 全代码
import requests
url = "http://item.jd.com/2967929.html"
try:
    r = requests.get(url)
    r.raise_for_status() #不返回200会产生异常
    r.encoding = r.apparent_encoding
    print(r.text[:1000])
except:
    print("爬取失败")
```
