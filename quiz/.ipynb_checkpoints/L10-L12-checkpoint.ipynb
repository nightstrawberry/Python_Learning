```python
#请使用列表推导式被小甲鱼不小心涂掉的部分
list1 = ['1.Just do it','2.一切皆有可能','4.Impossible is Nothing','3.让编程改变世界']
list2 = ['4.阿迪达斯','2.李宁','3.鱼C工作室','1.耐克']
#涂掉部分
list3 = [name[2:]+':'+slogan[2:] for name in list2 for slogan in list1 if name[0] == slogan[0]]
for each in list3:
    print(each)
```
