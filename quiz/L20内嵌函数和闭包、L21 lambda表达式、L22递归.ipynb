## 22 递归

```python
#非递归版本
def factorial(n):
    result = n
    for i in range(1,n):
        result *= i
    return result

number = int(input("请输入一个正整数:"))
result = factorial(number)
print("%d 的阶乘是 %d" %(number,result))
```

```python
#递归版本
def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)
    
factorial(4)
```

## L24 递归：汉诺塔 

```python
def hanoi(n,x,y,z):
    if n == 1:
        print(x,'->',z)
    else:
        hanoi(n-1,x,z,y) #将n-1个盘子从x移动到y上
        print(x,'->',z) #将最底下最后的盘子x到z上
        hanoi(n-1,y,x,z) #将y上n-1个盘子移动到z上
        
n = int(input('请输入汉诺塔层数：'))
hanoi(n,'x','y','z')
```

## L29:文件2 

任务：将文件按以下规律保存

1.甲鱼单独保存为boy_*.txt，客服单独保存为girl_*.txt（去掉甲鱼：、客服：）

2.文件三段对话保存为boy_1.txt,girl_1.txt,boy_2.txt,girl_2.txt...

```python
#打开目标文件
f = open('record.txt')
boy = []
girl = []
count = 1
#迭代文件每一行
#将字符串分割后存入boy[]、girl[]列表
for each_line in f:
    if each_line[:6] != '======':
        #分割字符串操作
        (role,line_spoken) = each_line.split('：',1)
        if role =='甲鱼':
            boy.append(line_spoken)
        if role =='客服':
            girl.append(line_spoken)
        #创建文件以存放分割后字符串
    elif each_line[:6] == '======'or each_line[:] == '':
        file_name_boy = 'boy_' + str(count) + '.txt'
        file_name_girl = 'girl_' + str(count) + '.txt'
        
        boy_file = open(file_name_boy,'w')
        girl_file = open(file_name_girl,'w')
        #写入字符串        
        #boy[]是列表，将之以序列形式写入
        boy_file.writelines(boy)
        boy_file.writelines(girl)
        #关闭文件
        boy_file.close()
        girl_file.close()
        #清空列表
        boy = []
        girl = []
        count += 1
   

f.close()
```

```python
#写一个程序，比较用户输入的两个文件，如果不同，显示所有不同不同字符行号和第一个不同字符的位置

def file_compare(file1,file2):
    f1 = open(file1)
    f2 = open(file2)
    count = 0 #统计行数
    differ = [] #统计不一样的数量
    
    for line1 in f1:
        line2 = f2.readline()
        count += 1
        if line1 != line2:
            differ.append(count)
            
    f1.close()
    f2.close()
    return differ

file1 = input('请输入需要比较的头一个文件名：')
file2 = input('请输入需要比较的另一个文件名：')

differ = file_compare(file1,file2)

if len(differ) == 0:
    print('一样！')
else:
    print('两个文件共有%d处不同'%len(differ))
    for each in differ:
        print('第%d行不一样'%each)
```
